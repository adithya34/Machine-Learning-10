## Load modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
type(iris)
```

```python
iris.data.shape
```

```python
iris_frame = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_frame.head()
```

## Dimensionality reduction to 3 features using PCA

```python
from sklearn.decomposition import PCA
PCA_data = PCA(n_components=3).fit_transform(iris.data)
```

## Data Visualization in a 3D plot and plot each class with a different color

```python
fig = plt.figure(figsize=(10, 8))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(PCA_data[:, 0], PCA_data[:, 1], PCA_data[:, 2], c=iris.target, s=50)
ax.set_title("PCA dimensions of transformed iris data")
ax.set_xlabel("First EigenVector")
ax.set_ylabel("Second EigenVector")
ax.set_zlabel("Third EigenVector")
plt.show()
```
